## Creating Class

- class ==>  vehicales
- attributes ==> color, model, make, price, brand
- methods ==> moveing, detail

```python
class vehicles():
    
    def __init__(self, color, model, price, brand): # initializer  (constructor)
        self.color = color
        self.model = model
        self.price = price
        self.brand = brand
        self.discount = 0.1 # static variables
    
    def details(self):
        print(f"brand : {self.brand}\
 model : {self.model}\
 color : {self.color}\
 price : {self.price}.")

    def moving(self):
        print(f"{self.brand} is a type of vehicle & can be used for travelling.")         
```

```python
toyota = vehicles("white", "Grande", 70, "toyota") # initialize
```

```python
print(type(toyota))
```

## Inheritance
- single parent single child Iheritance
- extra attirbute for single parnet single child inhertiance
- extra methods for single parnet single child inhertiance

```python
class parent():
    def __init__(self):
        self.haricolor = "brown"
        self.eyecolor = "brown"

    def property(self):
        print(f"Have a car")

class child(parent):
    def __init__(self, height):
        super().__init__()
        self.height = height

    def drive(self):
        print("Can drive a car.")
```

```python
shahood = child(6)
```

```python
shahood.eyecolor
```

```python
shahood.haricolor
```

```python
shahood.property()
```

```python
shahood.height
```

```python
shahood.drive()
```
